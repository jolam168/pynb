```python
root = "/Users/Lam/Downloads/logstash-7.2.0/data/output/raininfopro/"
file_name = root + "output.csv"
default_columns = ['Request_Time', 'url', 'url_path', 'url_querystring']
params = ['device_uuid', 'device_name', 'app_version', 'device_os']
result = default_columns + params

df = pd.read_csv(file_name, names=result, header=None)

df['Request_Time'] = pd.to_datetime(df['Request_Time'], format='%Y-%m-%dT%H:%M:%S')

display(df)

```

```python
import urllib.parse as urlparse
title = ['device_uuid','device_os','device_name','app_version']
def update_url(url, name):
    parsed = urlparse.urlparse(url)
    try: 
        return ''.join(urlparse.parse_qs(parsed.query)[name])
    except KeyError: 
        return np.nan


# df['ser'] = df['url_querystring'].apply(update_url, name=title[0])
# df['os_version'] =  df['url_querystring'].apply(update_url,name=title[1])
# df['device_name'] =  df['url_querystring'].apply(update_url,name=title[2])
# df['app_version'] =  df['url_querystring'].apply(update_url,name=title[3])
display(df.tail(20))
```
